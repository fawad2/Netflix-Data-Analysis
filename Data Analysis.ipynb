```python
#make a copy of dataset 
df = netflix.copy()
```

```python
#shape before removing null values
df.shape
```

```python
#dropping null values and checking shape
df = df.dropna()
df.shape
```

```python
#printing fist 10 values
df.head(10)
```

```python
#conver date time format
df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
sns.countplot(x = df["type"])

# Add a title and labels
plt.title("Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
ax = sns.countplot(x = df["type"])

# Add the percentage difference labeled under the chart
for i in ax.patches:
    height = i.get_height()
    ax.text(i.get_x() + i.get_width()/2, height + 2,
            str(np.round(height/df.shape[0]*100, 2)) + "%", ha="center")

# Add a title and labels
plt.title("Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
# Subset the data to only include rows where type is 'A' or 'B'
df_sub = df[df['type'].isin(['A', 'B'])]

# Create a dictionary to map the type column to a color
color_map = {'A': 'red', 'B': 'blue'}
colors = [color_map[x] for x in df_sub['type']]

# Create the packed bubble chart
plt.scatter(df_sub['type'], df_sub['value'], s=df_sub['count']*100, c=colors)

# Add a title and labels
plt.title("Type A vs Type B Distribution")
plt.xlabel("Type")
plt.ylabel("Value")

# Show the plot
plt.show()
```
